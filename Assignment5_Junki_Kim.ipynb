```python
# read a csv file into a dataframe
# got error when i try UTF-8 encoding
df_c = pd.read_csv('cuisines.csv', encoding="Windows-1252")
df_l = pd.read_csv('locations.csv', encoding="Windows-1252")
df_r = pd.read_csv('ratings.csv', encoding="Windows-1252")

print("Best rated restaurant need many requirement")
print("First is trust. It need to have clear information such as name and location, so people can trust.")
print("Second is famous. Restaurant which have many customer is already proved as good rated restaurant.")
print("Third is good customer review. Restaurant which have good customer review will be good rated restaurant.")
```

### 2. List the 10 best rated restaurants (Include their placeID, names and cuisine types) (2 points)

```python
df_c.describe() # 916 cuisines in data
df_l.describe() # 130 cuisines in data

df_cl = pd.merge(df_c, df_l, on='placeID', how ='inner')
df_cl.describe() # 112 cuisines has bothname and location

key_list = list(df_r['placeID'].unique())
key_list.sort()
key_list

data_list = []
for i in key_list:
    data_list.append(df_r[df_r['placeID']==i]['rating'].sum())
data_list

#sum of rating 
df_rsum = pd.DataFrame(data_list, key_list, columns=['sum'])

data_list = []
for i in key_list:
    data_list.append(df_r[df_r['placeID']==i]['rating'].mean())
data_list

#mean of rating
df_rmean = pd.DataFrame(data_list, key_list, columns=['mean'])

df_rmean.describe() # max mean is 2.00, min mean is 0.25

df_rmean[df_rmean['mean']==2.00] #132955 134986 135034
df_rmean[df_rmean['mean']==0.25] #132654 135040 has 0.25

df_toptenmean = df_rmean.sort_values(by=['mean']).tail(10)
df_toptensum = df_rsum.sort_values(by=['sum']).tail(10)

tempsum = df_toptensum.assign(placeID = df_toptensum.index)
tempmean = df_toptenmean.assign(placeID = df_toptenmean.index)

df_toptenm = pd.merge(df_cl, tempmean, on='placeID', how = 'inner')
df_toptens = pd.merge(df_cl, tempsum, on='placeID', how = 'inner')


#df_toptenm
```

#### 10 best rated restaurants sort by mean of rating value (good customer review)

```python
df_toptenm = df_toptenm.sort_values(by=['mean'], ascending=False)

print(df_toptenm.to_string(index=False))
```

#### 10 best rated restaurants sort by sum of rating value (many customer)

```python
df_toptens = df_toptens.sort_values(by=['sum'], ascending=False)

print(df_toptens.to_string(index=False))
```

### 4. Identify the most popular cuisine type in the top 10 list (1 point)
    - Note: it may not be the same as the most popular restaurantâ€™s cuisine type

```python
df_clr = pd.merge(df_cl, df_r, on='placeID', how ='inner')
df_clr

key_list = list(df_clr['cuisine'].unique())
key_list.sort()
key_list

data_list = []
for i in key_list:
    data_list.append(df_clr[df_clr['cuisine']==i]['rating'].sum())
data_list

df_ctsum = pd.DataFrame(data_list, key_list, columns=['sum'])

tempsum = df_ctsum.assign(cuisine = df_ctsum.index)
cols = ['cuisine', 'sum']

tempsum = tempsum[cols]
tempsum

df_cts = tempsum.sort_values(by=['sum'], ascending=False)
```

#### 10 most popular cuisine type sort by sum of rating value (many customer)
