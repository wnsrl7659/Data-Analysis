### 7. Create boxplots and compare them <u>pairwise</u> (4 points)
    - display boxplots in pairs (all possible pairs)
        - use subplots
        - compare:
            - sepal length betwwen all species
            - sepal width betwwen all species
            - petal width betwwen all species
            - petal length betwwen all species

```python
# Creating dataset 
fig, ax = plt.subplots(nrows=2,ncols=2, figsize=(10,8))

ax[0][0].set_title('sepal length', color="red")
ax[0][1].set_title('sepal width', color="green")
ax[1][0].set_title('petal length', color="red")
ax[1][1].set_title('petal width', color="green")

data_1 = mynparray[0:50,0:1]
data_1 = data_1.ravel()
data_2 = mynparray[50:100,0:1]
data_2 = data_2.ravel()
data_3 = mynparray[100:150,0:1]
data_3 = data_3.ravel()

data = [data_1, data_2, data_3]

ax[0][0].set_xticklabels(['setosa', 'versicolor', 'virginica']) 

ax[0][0].boxplot(data)

data_1 = mynparray[0:50,1:2]
data_1 = data_1.ravel()
data_2 = mynparray[50:100,1:2]
data_2 = data_2.ravel()
data_3 = mynparray[100:150,1:2]
data_3 = data_3.ravel()

data = [data_1, data_2, data_3]


ax[0][1].set_xticklabels(['setosa', 'versicolor', 'virginica']) 
ax[0][1].boxplot(data)

data_1 = mynparray[0:50,2:3]
data_1 = data_1.ravel()
data_2 = mynparray[50:100,2:3]
data_2 = data_2.ravel()
data_3 = mynparray[100:150,2:3]
data_3 = data_3.ravel()

data = [data_1, data_2, data_3]


ax[1][0].set_xticklabels(['setosa', 'versicolor', 'virginica']) 
ax[1][0].boxplot(data)

data_1 = mynparray[0:50,3:4]
data_1 = data_1.ravel()
data_2 = mynparray[50:100,3:4]
data_2 = data_2.ravel()
data_3 = mynparray[100:150,3:4]
data_3 = data_3.ravel()

data = [data_1, data_2, data_3]


ax[1][1].set_xticklabels(['setosa', 'versicolor', 'virginica']) 
ax[1][1].boxplot(data)
  
# show plot 
plt.show() 
```
